```python
import scipy
from matplotlib import pyplot as plt
import numpy as np
```

```python
from scipy.io.wavfile import write #a function that can save wave files

T = 1 # the number of seconds to simulate
samplerate = 44100 #the default sample rate for audio files

time = np.linspace(0, T, T*samplerate, dtype = np.float32) #a grid of points spanning 0,T with T*samplerate divisions 
omega  = 2 * np.pi * 261.625565 #the frequency in hertz of middle C the musical note
wave = np.sin(omega * time) # a simple sine wave at that frequency

write("example.wav", samplerate, wave)

figure, axis = plt.subplots(figsize = (20,5))
axis.plot(time, wave)
axis.set(xlim = (0,0.5), xlabel = 'time (s)') #only plot the first 10 miliseconds
```

```python
T = 10 # the number of seconds to simulate
samplerate = 44100 #the default sample rate for audio files

time = np.linspace(0, T, T*samplerate, dtype = np.float32) #a grid of points spanning 0,T with T*samplerate divisions 
omega  = 2 * np.pi * 261.625565 #the frequency in hertz of middle C the musical note

#let's change the pitch by 5% at 5hz
modulation = 0.05 * omega * np.sin(2 * np.pi * 5 * time)

modulate_carrier_wave = np.sin((omega + modulation) * time) # a simple sine wave at that frequency

write("modulated_carrier_wave.wav", samplerate, modulate_carrier_wave)

figure, axes = plt.subplots(2,1, figsize = (20,10), sharex = 'all')
axes[0].plot(time, modulation)
axes[1].plot(time, modulate_carrier_wave)
axis.set(xlim = (0,1), xlabel = 'time (s)') #only plot some of the time
```
